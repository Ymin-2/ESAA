#**암환자 유전체 데이터의 변이 정보를 활용한 암종 분류 AI 모델 개발**

##**데이터**

train.csv
- 총 6201개의 행(rows)
- ID : 샘플 별 고유 ID
- SUBCLASS : 암종 (26개 존재)
- 암환자 유전체 변이 정보 데이터 ( 유전체 관련 컬럼 4384개 )

test.csv
- 총 2546개의 행(rows)
- ID : 샘플 별 고유 ID
- 암환자 유전체 변이 정보 데이터 ( 유전체 관련 컬럼 4384개, train.csv와 동일 )

sample_submission.csv
- ID : 샘플 별 고유 ID
- SUBCLASS : 예측한 암종 (학습 데이터와 동일한 26개 존재)


##**데이터 분석**

1. 데이터 개요
  - 데이터는 HGVS 형태의 돌연변이 데이터
  - 각 유전자(열)에 대해 샘플(행)이 어떤 돌연변이를 가지는지 나타냄
  - 돌연변이의 종류
2. 데이터 분석
  - 특정 유전자는 돌연변이가 많이 일어남
  - 데이터 불균형이 꽤 있음
  - 암종별로 주로 돌연변이가 일어나는 유전자가 다름
3. 전처리
  - 행별로 [변한 아미노산의 종류] / [돌연변이의 종류]에 따른 변이 개수를 합한 값
  - [돌연변이 종류]별로 가중치를 부여하여 합산
  - 여러가지 행별 피처 추가
4. 외부 데이터 도입 및 처리
  - 주어진 데이터만으로는 성능 향상에 한계가 있음을 판단
  - train 데이터와 유사하게 분류된 TCGA Dataset 발견
  - TCGA 데이터 중 data_mutations_[암종].txt 내의 HGVSp_Shot 열 사용
  - 데이터에 오류가 있는 부분 제거 및 훈련에 필요한 변이만 추출
  - train에 있지 않은 유전자에 대한 변이 무시
5. 모델 구축 및 검증
  - 메인 모델: LGBM with Focal loss
    - 데이터의 클래스 불균형 문제를 다루기 위해 Focal loss 도입
  - 서브 모델: LGBM
    - 메인 모델이 KIPAN - KIRC / LGG - GBMLGG의 관계를 잘 파악하지 못함
  - 전체 모델
    - 클래스 불균형이 있기 때문에 클래스 비율을 유지하며 validation data 구축

##**차별점 및 배울점**

- 도메인 확장 능력: 내부 데이터 한계를 파악하고 TCGA 같은 외부 공공데이터를 연동
- 모델 세트 구상 능력: 한 모델이 구조적 관계(KIPAN-KIRC / LGG-GBMLGG)를 못 잡는 걸 발견 → 보조 모델 추가 → 전체 조합
- 외부 데이터를 train 구조에 맞게 변환
- 데이터 불균형에 대한 명확한 대응: Focal Loss + class 비율 유지한 validation
