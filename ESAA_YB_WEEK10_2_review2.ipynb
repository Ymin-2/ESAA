#**주제**
난임 환자 대상 임신 성공 확률 예측 AI모델 개발

##**데이터**

* train.csv
  - ID : 샘플별 고유 ID (126,244 rows)
  - 난임 환자 시술 데이터 (33개의 컬럼)
  - 임신 성공 확률 : 해당 주기의 난임 시술로부터 임신에 성공할 확률 (0~1)

* test.csv
  - ID : 샘플별 고유 ID (54,412 rows)
  - 난임 환자 시술 데이터 (33개의 컬럼)

* sample_submission.csv
  - ID : 샘플별 고유 ID
  - 임신 성공 확률 : 해당 주기의 난임 시술로부터 예측한 임신에 성공할 확률 (0~1)

##**코드 리뷰**

1. 데이터 로드 이후 칼럼 분류
  - categorical, numeric, bool형 칼럼 리스트 미리 지정
2. 커스텀 평가 지표 정의
  - f1_score: 확률을 0.5 기준으로 0/1로 바꿔서 F1 점수 계산
  - weighted_brier_score: 예측 오차에 가중치 적용 후 1 - brier 형태로 반환
  - competition_metric: weighted Brier score와 F1 점수를 절반씩 섞은 최종 metric
3. 결측치 처리
  - 시술 유형이 'DI'이거나 이식된 배아 수가 0인 경우 결측치를 0으로 채움
4. 특정 범주값 정제
  - '알 수 없음' 값을 특정 범주로 치환
5. DI 시술 데이터 전처리 함수
  - DI 시술 행의 일부 결측 수치를 0으로 대체
6. 결측치 처리
  - 특정 시술 유형(DI, IVF)일 때 결측치를 0으로 채움
7, 수치형 범주 매핑
  - first(df) 정의: 문자열로 된 횟수 범주를 숫자값으로 변환, 범위형 구간도 대표값으로 치환
8. 컬럼 타입 업데이트 함수(unpdate_column_types_with_drop)
  - drop_list에 있는 칼럼 제거
  - 남은 칼럼 중 숫자/문자 타입 자동 분류해 리스트 갱신
9. 인코딩 및 결측치 처리
  - OrdinalEncoder로 범주를 숫자형으로 변환
  - 평균 대체로 결측치 채움
10. 상관관계 시각화
11. 학습용 데이터 분할
  - 8:2로 train/test 세트 분리
12. PyTorch Dataset 정의
  - 테이블 데이터를 PyTorch 텐서로 변환해 DataLoader에 쓸 수 있게 함
13. MLP 모델 정의
14. 가중 브라이어 손실 함수
  - 정답값 중심으로 가중치를 주는 Brier loss
15. 학습 루프 + 상위 모델 저장
16. 학습 실행
17. 앙상블 예측
18. 결과 저장
19. 두 번째 동일 학습 반복
20. 최종 앙상블
